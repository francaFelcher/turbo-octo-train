# 2. Read .csv file from Multiple URLs with Pandas

https://www.football-data.co.uk/mmz4281/2425/E0.csv
### Link: root + season + league

```python
# link structure
"https://www.football-data.co.uk/mmz4281" + "2425" + "/" + "E0" + ".csv"
```

```python
# create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

#### MULTIPLE LEAGUES

```python
leagues = ["E0","E1","E2","E3","EC"]
frames = []

# loop through leagues, read multiple csv and append it into a list
for league in leagues:
    df = pd.read_csv(root + "2425" + "/" + league + ".csv") # se non funziona, aggiungere: , encoding = "unicode_escape"
    frames.append(df)
```

#### MULTIPLE SEASON

```python
for season in range(15, 24):
    print(str(season) + str(season + 1))
```

```python
leagues = ["E0","E1","E2","E3","EC"]
frames = []

# loop through leagues, read multiple csv and append it into a list
for league in leagues:
    for season in range(20, 24):
        df = pd.read_csv(root + (str(season) + str(season + 1)) 
                         + "/" + league + ".csv", encoding = "unicode_escape") # se non funziona, aggiungere: , encoding = "unicode_escape"
        df.insert(1, "season", ("20"+str(season)))
        frames.append(df)
```

# 3. Organize all the Data in a Dictionary

```python
# create a dictionary with original name of league as key
dict_countries = {"Spanish La Liga":"SP1",
                 "Spanish Segunda Division":"SP2",
                 "German Bundesliga":"D1",
                 "English Premier League":"E0",
                 "English League 1":"E2"}
```

```python
# get a dictionary element
dict_countries["Spanish La Liga"]
```

```python
for league in dict_countries:
    print(dict_countries[league])
```

```python
dict_historical_data = {}

# leagues = ["E0","E1","E2"]


# loop through key elements
for league in dict_countries:
    frames = []
    for season in range(20, 24):
        df = pd.read_csv(root + (str(season) + str(season + 1)) 
                         + "/" + dict_countries[league] + ".csv", 
                         encoding = "unicode_escape") # se non funziona, aggiungere: , encoding = "unicode_escape"
        df.insert(1, "season", ("20"+str(season)))
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```
