```python
df_population_raw = pd.read_csv("population_total.csv")
df_population_raw.dropna(inplace=True)
df_pivot=df_population_raw.pivot(index="year",
                                columns="country",
                                values="population")
df_pivot=df_pivot[["United States", "India", "China", "Indonesia", "Brazil"]]
df_pivot
```
