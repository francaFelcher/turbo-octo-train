```python
import pandas as pd
```

```python
df_laptops = pd.read_csv("laptop_price.csv", encoding="unicode_escape")
```

```python
df_laptops.head()
```

# drop_duplicates()

```python
df_laptops.drop_duplicates(["Company"])[["Company", "Price_euros"]]
```

```python
# sort df ascending by company and price
# cheapest first and most expansive last
df_laptops = df_laptops.sort_values(["Company", "Price_euros"])
```

```python
# cheapest: keep="first"
df_laptops.drop_duplicates(["Company"], keep="first")[["Company","Price_euros"]]
```

```python
# most expansive: keep="last"
df_laptops.drop_duplicates(["Company"], keep="last", inplace=True, ignore_index=True)
```

Task: find the biggest and smallest screen size in laptops of each company using the
sort_values() and duplicated() methods (keep "first", "last" and False)

```python
# sort df ascending by company and inches
# smallest first and biggest last
df_laptops = df_laptops.sort_values(["Company","Inches"])
```

```python
# smallest: keep="first"
    # duplicated_first = df_laptops.duplicated("Company", keep="first")
    # df_laptops[~duplicated_first]
df_laptops.drop_duplicates("Company", keep="first")[["Company","Inches"]]
```

```python
# biggest: keep="last"
    # duplicated_last = df_laptops.duplicated("Company", keep="last")
    # df_laptops[~duplicated_last]
df_laptops.drop_duplicates("Company", keep="last")[["Company", "Inches"]]
```
