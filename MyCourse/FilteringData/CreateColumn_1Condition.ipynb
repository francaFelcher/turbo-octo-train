# np.where (1 Column, 1 Condition)

```python
import numpy as np
```

```python
tier_list = np.where(df_laptops["Price_euros"]>2000, "Expansive", "Cheap")
```

```python
df_laptops["Price_tier"] = tier_list
```

```python
df_laptops
```

```python
df_laptops.value_counts("Price_tier")
```
