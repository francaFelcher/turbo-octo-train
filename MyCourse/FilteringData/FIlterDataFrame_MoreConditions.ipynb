```python
import pandas as pd
```

```python
df_laptops = pd.read_csv("laptop_price.csv", encoding="unicode_escape")
```

```python
df_laptops.head()
```

# Filter: 2 or more Conidition

```python
# find apple laptops that cost more than 2000
(df_laptops["Company"]=="Apple") & (df_laptops["Price_euros"]>2000)
```

#### & (and)

```python
df_laptops[(df_laptops["Company"]=="Apple") & (df_laptops["Price_euros"]>2000)]
```

```python
df_laptops.value_counts((df_laptops["Company"]=="Apple") & (df_laptops["Price_euros"]>2000))
```

#### | (or)

```python
df_laptops.head()
```

```python
# find laptops Apple or Dell
(df_laptops["Company"]=="Apple") | (df_laptops["Company"]=="Dell")
```

```python
df_laptops[(df_laptops["Company"]=="Apple") | (df_laptops["Company"]=="Dell")]
```

```python
df_laptops[(df_laptops["Company"]=="Apple") | (df_laptops["Company"]=="Dell")].value_counts("Company")
```
