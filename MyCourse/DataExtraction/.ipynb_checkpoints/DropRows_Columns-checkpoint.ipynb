```python
import pandas as pd
```

```python
df = pd.read_csv("players_20.csv")
df.set_index("short_name", inplace = True)
df = df[["long_name","age","dob","height_cm","weight_kg","nationality","club"]]
```

```python
df.head(3)
```

# Drop Row

```python
# axis parameter
df.drop("L. Messi", axis = 0)
# index parameter
df.drop(index=["L. Messi"])
```

```python
# drop two or more rows and update data
df.drop(["L. Messi","Cristiano Ronaldo"], axis = 0, inplace = True)
```

```python
df.head(3)
```

# Drop columns

```python
df.drop("long_name", axis = 1)

#column parameter
df.drop(columns=["long_name"])
```

```python
# drop column by position/index (drop last columns)
df.columns[[-1]]
df.drop(df.columns[[-1]], axis = 1)
```

```python
# drop two or more columns and update
df.drop(["long_name","dob"], axis=1, inplace=True)
```

```python
df.head(3)
```
