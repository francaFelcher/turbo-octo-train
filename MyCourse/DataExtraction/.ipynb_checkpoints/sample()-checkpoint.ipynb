```python
import pandas as pd
```

```python
df = pd.read_csv("players_20.csv")
df.set_index("short_name", inplace = True)
df = df[["long_name","age","dob","height_cm","weight_kg","nationality","club"]]
```

```python
df.head(3)
```

# sample()

```python
# extract 10 random elements from "nationality" column
df.sample(5).value_counts("nationality")
```

```python
df["nationality"].sample(10, random_state=99)
```

```python
# extract a random 20% sample of the dataframe
df.sample(frac=0.2, random_state=2)
```

```python
# upsample : increase the sampling rate -> frac>1
df.sample(frac=2, replace=True, random_state=12)
```
