```python
import pandas as pd
```

```python
df_exams = pd.read_csv("StudentsPerformance.csv")
df_exams.head()
```

# Create Index

```python
import numpy as np
import random
```

```python
# non repetative values for the index
new_index = np.arange(0, 1000)
```

```python
# shuffling indexes
random.shuffle(new_index)
```

```python
# create a column with new index
# ( to remove a column:    df_exams.drop("new index", axis='columns', inplace=True)    )
df_exams.insert(0, "new index", new_index)
```

# Set Index

```python
# set new_index column as index
df_exams.set_index("new index", inplace = True)
```

```python
df_exams
```
