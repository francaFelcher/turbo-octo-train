```python
import pandas as pd
```

```python
df_exams = pd.read_csv("StudentsPerformance.csv")
df_exams.head()
```

# Rename Column

```python
# rename column and overwrite values of the dataframe
df_exams = df_exams.rename(columns={"gender" : "Gender"})
```

```python
df_exams.head()
```

```python
# rename multiple columns and update the datagrame with inplace argument
df_exams.rename(columns={"math score":"M_S","reading score":"R_S","writing score":"W_S"}, inplace = True)
```

```python
df_exams
```

# Rename Index

```python
# rename index 0, 1, 2 and update the dataframe
df_exams.rename(index = {0:"A", 1: "B", 2: "C"}, inplace = True)
```
