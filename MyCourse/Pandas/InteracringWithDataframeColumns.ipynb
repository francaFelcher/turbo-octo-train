### ADDING A NEW COLUMN WITH assign() AND insert()

#### assign()
When to use?
- add multiples columns in a single line of code
- overwrite the values of an existing column (best practice)

```python
score1 = np.random.randint(1, 101, size=1000)
score2 = np.random.randint(1, 101, size=1000)
```

```python
series1 = pd.Series(score1, index=np.arange(0,1000))
series2 = pd.Series(score2, index=np.arange(0,1000))
```

```python
series1
```

```python
series2
```

```python
df_exams = df_exams.assign(score1 = series1, score2 = series2)
```

#### insert()
Insert a new column at a specific position or index

```python
df_exams.insert(1, "test", series1)
```

```python
df_exams
```
